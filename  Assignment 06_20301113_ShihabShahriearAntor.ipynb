```python
#problem 02

class Assassin:
    num_assassin = 0
    def __init__(self, name, s_rate):
        self.name = name
        self.s_rate = s_rate
        
    @classmethod
    def failureRate(cls, name, f_rate):
        return cls(name, 100-f_rate)
        
       
    @classmethod
    def failurePercentage(cls, name, fP):
        return cls(name, 100-fP)
        
    def printDetails(self):
        Assassin.num_assassin = Assassin.num_assassin + 1
        print("Name:", self.name)
        print("Success rate:", self.s_rate)
        print("Total number of Assassin:", Assassin.num_assassin)



john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#problem 03

class Passenger:
    count = 0
    def __init__(self, name):
        self.name = name
        
    def set_bag_weight(self, bag_weight):
        
        if bag_weight <= 20:
            self.bus_fare = 450
        elif bag_weight >= 21 and bag_weight <= 50:
            self.bus_fare = 450 + 50
        else:
            self.bus_fare = 450 + 100
            
        
    def printDetail(self):
        Passenger.count = Passenger.count + 1
        print("Name:", self.name)
        print("Bus Fare:", self.bus_fare)


print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
#problem 07

class Cat:
    
    Number_of_cats = 0

    def __init__(self, color = "White", position = "sitting"):
        self.color = color
        self.position = position
        Cat.Number_of_cats = Cat.Number_of_cats + 1
        
    @classmethod
    def no_parameter(cls):
        return cls()
    @classmethod
    def first_parameter(cls, color):
        return cls(color)
    @classmethod
    def second_parameter(cls, pos):
        return cls("Grey", pos)
    
    def changeColor(self, color):
        self.color = color
    def printCat(self):
        print(f"{self.color} cat is {self.position}")
        

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```
